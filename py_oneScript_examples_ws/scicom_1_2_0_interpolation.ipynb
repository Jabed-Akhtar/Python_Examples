# Interpolation - Linear and Cubic Splines
Based on the discrete data points, interpolation is a method to find the new data points.

### Types
- Piecewise constant interpolation
- Linear interpolation
- Polynomial interpolation
- Spline interpolation
- Mimetic interpolation

### Usages of Interpolation
Interpolation, the mathematical technique for estimating values between known data points, finds widespread applications across various fields. Its versatility makes it a valuable tool for enhancing data representation, analysis, and modeling.

#### Common Usages:

1. **Data Visualization:**
   - Interpolation is used in graphics and data visualization to create smooth curves or surfaces that represent trends or patterns in the data. This enhances the visual representation of data.

2. **Image Processing:**
   - In image processing, interpolation is employed to resize images, enhance resolution, or generate intermediate pixels when transforming images. Common interpolation methods include bilinear and bicubic interpolation.

3. **Digital Signal Processing:**
   - In signal processing, interpolation is used to upsample signals, reconstruct missing data points, or interpolate between samples. This is crucial in applications like audio processing and communication systems.

4. **Geographic Information Systems (GIS):**
   - GIS applications use interpolation to estimate values of geographic features between known data points. This is useful in creating continuous elevation models, temperature maps, or other spatial representations.

5. **Scientific Computing:**
   - Interpolation is a fundamental tool in scientific computing for estimating values of functions at points not explicitly given in the dataset. It is commonly used in numerical simulations and modeling.

6. **Control Systems:**
   - In control systems engineering, interpolation is used to estimate intermediate values between discrete measurements. This is particularly important for smooth control signals and trajectory planning.

7. **Navigation and GPS:**
   - GPS systems often use interpolation to estimate the position of a moving object between discrete GPS measurements. This improves the accuracy of tracking and navigation.

8. **Finance and Economics:**
   - Interpolation is employed in finance for estimating values between known market data points. It is used in curve fitting for interest rate curves, yield curves, and other financial instruments.

9. **Medical Imaging:**
   - In medical imaging, interpolation techniques are used to reconstruct images with higher resolution or to generate slices between existing images. This is important for diagnostic imaging techniques like CT scans and MRI.

10. **Computer-Aided Design (CAD):**
    - CAD applications use interpolation to create smooth curves and surfaces between specified control points. This is essential for designing complex shapes and structures.

11. **Animation and Computer Graphics:**
    - Interpolation is extensively used in animation and computer graphics to create smooth motion and transitions between keyframes. It helps generate realistic movement in virtual environments.

12. **Machine Learning and Data Analysis:**
    - Interpolation can be used to fill in missing data points in datasets, allowing for more complete analyses. It is also employed in preprocessing data for machine learning algorithms.

13. **Environmental Modeling:**
    - Interpolation is applied in environmental science for modeling and predicting phenomena like pollution levels, temperature distribution, and precipitation across a geographic area.

These examples highlight the versatility of interpolation and its widespread use in diverse fields to enhance data representation, analysis, and modeling.

### Usages of Interpolation in Control System
Interpolation plays a crucial role in control systems engineering, providing valuable tools for estimating and generating intermediate values between known data points. Here are common usages of interpolation in control systems:

#### 1. Sensor Data Processing:

In control systems, sensors often provide discrete measurements at specific time instances. Interpolation is used to estimate values of signals between these discrete measurements, providing a continuous representation of the system's state.

#### 2. Trajectory Planning:

Interpolation is employed in robotics and motion control for planning trajectories smoothly between key waypoints. This ensures that the motion of a robotic system is continuous and free of abrupt changes, reducing wear and tear.

#### 3. Control Signal Generation:

Interpolation is used to generate intermediate control signals between consecutive time steps. This is valuable in systems where control inputs need to be smoothly varied to avoid sudden changes in the controlled variables.

#### 4. State Estimation:

Control systems often involve estimating the state of a dynamic system. Interpolation aids in estimating the continuous evolution of the system state between discrete measurements, enhancing the accuracy of state estimation.

#### 5. Signal Reconstruction:

In cases where signals are sampled at discrete intervals, interpolation is used to reconstruct a continuous representation of the signal. This is particularly important in control applications that involve feedback from sensors.

These applications demonstrate how interpolation contributes to the smooth operation, accurate representation, and optimal performance of control systems. It allows engineers to bridge the gap between discrete data points, facilitating continuous and efficient operation.

### Example#1: Control System - Sensor Data Processing

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate sensor data with noise
time = np.linspace(0, 10, 100)
true_signal = np.sin(time)
noisy_sensor_data = true_signal + 0.1 * np.random.randn(len(time))

# Interpolate sensor data for smoother representation
interp_time = np.linspace(0, 10, 1000)
interp_sensor_data = np.interp(interp_time, time, noisy_sensor_data)

# Plot the results
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(time, noisy_sensor_data, 'o', label='Noisy Sensor Data')
plt.title('Original Noisy Sensor Data')
plt.xlabel('Time')
plt.ylabel('Sensor Reading')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(interp_time, interp_sensor_data, label='Interpolated Sensor Data')
plt.title('Interpolated Sensor Data')
plt.xlabel('Time')
plt.ylabel('Sensor Reading')
plt.legend()

plt.tight_layout()
plt.show()
```

### Example#2: Control System - Trajectory Planning

```python
import numpy as np
import matplotlib.pyplot as plt

# Waypoints
waypoints = np.array([[0, 0], [1, 2], [3, 1], [5, 4]])

# Interpolate trajectory between waypoints
t = np.linspace(0, 1, 100)
x_interp = np.linspace(0, 5, 15)
waypoint_interp = np.interp(x_interp, waypoints[:, 0], waypoints[:, 1])
#trajectory_x = np.interp(t, [0, 1], waypoints[:, 0])
#trajectory_y = np.interp(t, [0, 1], waypoints[:, 1])

# Plot the results
plt.figure(figsize=(6, 6))
plt.plot(waypoints[:, 0], waypoints[:, 1], 'ro', label='Waypoints')
plt.plot(x_interp, waypoint_interp, '*', label='Waypoints-interp')
plt.title('Trajectory Planning')
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.legend()
plt.grid(True)
plt.show()
```

### Example#3: Control System - Control Signal Generation

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a control signal (e.g., for a motor)
time = np.linspace(0, 10, 100)
control_signal = np.sin(time) + 0.2 * np.random.randn(len(time))

# Interpolate control signal for smoother representation
interp_time = np.linspace(0, 10, 1000)
interp_control_signal = np.interp(interp_time, time, control_signal)

# Plot the results
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(time, control_signal, label='Control Signal')
plt.title('Original Control Signal')
plt.xlabel('Time')
plt.ylabel('Control Value')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(interp_time, interp_control_signal, label='Interpolated Control Signal')
plt.title('Interpolated Control Signal')
plt.xlabel('Time')
plt.ylabel('Control Value')
plt.legend()

plt.tight_layout()
plt.show()
```

### Example#4: Control System - State Estimation

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate true system state with noise
time = np.linspace(0, 10, 100)
true_state = np.sin(time)
noisy_state = true_state + 0.1 * np.random.randn(len(time))

# Interpolate state for smoother representation
interp_time = np.linspace(0, 10, 1000)
interp_state = np.interp(interp_time, time, noisy_state)

# Plot the results
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(time, noisy_state, 'o', label='Noisy State')
plt.title('Original Noisy System State')
plt.xlabel('Time')
plt.ylabel('System State')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(interp_time, interp_state, label='Interpolated System State')
plt.title('Interpolated System State')
plt.xlabel('Time')
plt.ylabel('System State')
plt.legend()

plt.tight_layout()
plt.show()
```
