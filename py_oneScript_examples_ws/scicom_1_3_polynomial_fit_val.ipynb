# Polynomial Fitting and Evaluation
**Polynomial Fitting:**
- Polynomial Regression
- Curve Fitting
- Least Squares Fitting
- Polynomial Approximation

**Polynomial Evaluation:**
- Polynomial Interpolation (when used with known points)
- Polynomial Prediction
- Polynomial Function Evaluation

These names reflect the nature of the operations. Polynomial fitting involves finding the best-fitting polynomial to a set of data points, often using least squares methods, while polynomial evaluation involves calculating the values of a polynomial for given input points.

Remember that the specific names used might vary depending on the context and the software/documentation you are working with. In the context of Python and NumPy/SciPy, "polyfit" and "polyval" are widely accepted terms.

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 1, 4, 3, 5])

# Degree of the polynomial fit
degree = 2

# Polynomial fitting using polyfit
coefficients = np.polyfit(x, y, degree)

# Polynomial evaluation using polyval
x_values = np.linspace(min(x), max(x), 100)
y_values = np.polyval(coefficients, x_values)

# Plotting the original points and the fitted polynomial
plt.scatter(x, y, label='Original Points')
plt.plot(x_values, y_values, label=f'Polynomial Fit (degree={degree})', color='red')
plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Polynomial Fitting and Evaluation Example')
plt.show()
```
