## Usage of Taylor Series in **Control System**
In control systems engineering, the Taylor series and its applications are commonly used to linearize nonlinear systems. Linearization is a technique that simplifies the analysis and design of control systems by approximating a nonlinear system with a linear one around a specific operating point or equilibrium condition. This is particularly useful when dealing with complex systems where analytical solutions may be difficult to obtain.

### Linearization using Taylor Series:  
The Taylor series expansion of a function f(x) around a point a is given by:  
<img src="notebook_imgs/taylor_fun.png" alt="Example Image" width="500"/>  
In control systems, this expansion is often truncated after the first few terms for simplicity. Linearization involves selecting an operating point x=a and considering only the terms up to the first-order derivative, resulting in a linear approximation.

### Detailed Steps for Linearization

1. **Select Operating Point:** Identify the desired operating point for the system.
2. **Define State and Input Variables:** Define the state variables \(x\) and control input variables \(u\) for the system.
3. **Compute Partial Derivatives:** Compute the partial derivatives of the nonlinear dynamics \(f(x, u)\) with respect to \(x\) and \(u\) at the operating point.
4. **Form Matrices A and B:** Use the partial derivatives to form matrices \(A\) and \(B\) in the linearized system representation \(\dot{x} = Ax + Bu\).
5. **Apply Control Techniques:** Apply linear control techniques and tools to design controllers, analyze stability, and evaluate system performance.

### Example#1: Function Approximation

Let's consider the function f(x) = sin(x) and use a Taylor series expansion to approximate it around a specific point. In this example, we'll approximate sin(x) around x = 0 using the first few terms of the Taylor series.

In the code, we use the Taylor series expansion for \(\sin(x)\):

\[
\sin(x) \approx x - \frac{x^3}{3!} + \frac{x^5}{5!} - \ldots
\]

<img src="notebook_imgs/taylor_sine.png" alt="Example Image" width="200"/>

The function `original_function` represents the true \(\sin(x)\), and `taylor_approximation` computes the Taylor series approximation up to the specified order. The code then plots the original function along with the Taylor approximations for orders 1, 3, and 5.

```python
# Imports
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt

def original_function(x):
    return np.sin(x)

def taylor_approximation(x, n):
    # Compute the terms of the Taylor series expansion for sin(x) around x=0
    terms = [((-1)**i) * x**(2*i + 1) / np.math.factorial(2*i + 1) for i in range(n)]
    return np.sum(terms, axis=0)

# Values for x
x_values = np.linspace(-2*np.pi, 2*np.pi, 100)

# Plot the original function
plt.figure(figsize=(10, 6))
plt.plot(x_values, original_function(x_values), label='sin(x)', color='blue')

# Plot Taylor approximations for different orders
for order in [1, 2, 3, 5]:
    approximation_values = taylor_approximation(x_values, order)
    plt.plot(x_values, approximation_values, label=f'Taylor (Order {order})', linestyle='dashed')

plt.title('Function Approximation with Taylor Series')
plt.xlabel('x')
plt.ylabel('Function Value')
plt.ylim([-7.5, 7.5])
plt.legend()
plt.grid(True)
plt.show()
```

### Example: Linearization using Taylor Series
Let's consider a simple nonlinear system:  
x_dot = x^2 + u

Linearizing around an operation point (x_a, u_a) = (0, 0), the linearized system is:  
x_dot ≈ u

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Nonlinear system dynamics
def nonlinear_system(x, t, u):
    return x**2 + u

# Linearized system dynamics
def linearized_system(x, t, u):
    return u

# Simulation
t = np.linspace(0, 5, 100)
u_input = np.sin(t)  # Example input signal

# Nonlinear simulation
x_nonlinear = odeint(nonlinear_system, 0, t, args=(u_input,))

# Linearized simulation
x_linearized = odeint(linearized_system, 0, t, args=(u_input,))

# Plot results
plt.figure(figsize=(10, 6))
plt.plot(t, x_nonlinear, label='Nonlinear System')
plt.plot(t, x_linearized, label='Linearized System')
plt.xlabel('Time')
plt.ylabel('State (x)')
plt.legend()
plt.grid(True)
plt.title('Comparison of Nonlinear and Linearized Systems')
plt.show()
```
