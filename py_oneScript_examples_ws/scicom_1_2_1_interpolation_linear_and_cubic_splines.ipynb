# Interpolation - Linear and Cubic Splines
### Linear Spline (Explained with example)
Suppose you have a set points ***(x0, y0), (x1, y1), (x2, y2), ..., (xn, yn)*** that represents waypoints in a __trajectory__. A linear spline between two consecutive points ***(x0, y0)*** and ***(xi+1, y1+1)*** can be expressed as:  
y(x) = m_i * (x - x_i) + b_i  
where m_i is the slope and b_i is the y-intercept for the linear segment betwn ***(x0, y0)*** and ***(xi+1, y1+1)***. The slope m_i is calculated as:  
m_i = (yi+1 - yi) / (xi+1 - xi)

### Example: Control System - Trajectory Planning

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

# Sample data for trajectory points
t = np.array([0, 1, 2, 3, 4])
x = np.array([0, 1, 3, 4, 6])
y = np.array([0, 2, 1, 3, 2])

# Linear spline interpolation
linear_interp = interp1d(t, np.column_stack((x, y)), kind='linear', axis=0)

# Cubic spline interpolation
cubic_interp = interp1d(t, np.column_stack((x, y)), kind='cubic', axis=0)

# Generating finer time points for smoother trajectory visualization
t_fine = np.linspace(t[0], t[-1], 100)

# Interpolating the trajectory using linear and cubic splines
trajectory_linear = linear_interp(t_fine)
trajectory_cubic = cubic_interp(t_fine)

# Plotting the original points and the interpolated trajectories
plt.scatter(x, y, label='Original Points', color='red')
plt.plot(trajectory_linear[:, 0], trajectory_linear[:, 1], label='Linear Spline', linestyle='dashed', color='blue')
plt.plot(trajectory_cubic[:, 0], trajectory_cubic[:, 1], label='Cubic Spline', color='green')
plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Linear and Cubic Spline Interpolation for Trajectory Planning')
plt.show()
```
