## Explained with an Example - Vehicle's Longitudinal Speed Control
**Tags:** ```simpson rule``` ```quadratic polynomial``` ```cumulative sum```

Let's assume we have a numpy array ***time*** representing time points and another array ***velocity*** representing the corresponding velocity at each time point.

#### Simpson's rule
Simpson's rule is a numerical integration technique that approximates the definite integral of a function by using quadratic polynomials. The formula for Simpson's rule is as follows:  
<img src="notebook_imgs/simps_fun.png" alt="Example Image" width="500"/>

````python
import numpy as np
from scipy.integrate import simps
import matplotlib.pyplot as plt

# Generate sample data
time = np.linspace(0, 10, 100)  # Time points from 0 to 10 seconds
velocity = 5 + 2 * time  # A simple linear velocity profile (you can replace this with your actual data)

# Integrate velocity to get distance traveled
distance_traveled = simps(velocity, time)

print(f"Total distance traveled: {distance_traveled:.2f} meters")

# Plot the velocity profile
plt.plot(time, velocity, label='Velocity Profile')
plt.fill_between(time, velocity, alpha=0.2, label='Area under the curve')

plt.xlabel('Time (seconds)')
plt.ylabel('Velocity (m/s)')
plt.title('Vehicle Longitudinal Speed Control')
plt.legend()
plt.show()
````

### Numerical Integration
To find the distance travelled (delta(x)) over a time interval (delta(t)), we can integrate the velocity profile with respect to time:  
<img src="notebook_imgs/dist_velo_numerical_fun.png" alt="Example Image" width="100"/>  
In a discrete form using numerial integration, we can use cumulative sum of the velocity multiple by the time step:  
<img src="notebook_imgs/dist_velo_sum_fun.png" alt="Example Image" width="150"/>

````python
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data
time = np.linspace(0, 10, 100)
velocity = 5 + 2 * time

# Calculate the distance traveled using numerical integration
delta_time = np.diff(time)  # Time step
distance_traveled = np.cumsum(velocity[:-1] * delta_time)

# Print the total distance traveled
print(f"Total distance traveled: {distance_traveled[-1]:.2f} meters")

# Plot the velocity profile
plt.subplot(2, 1, 1)
plt.plot(time, velocity, label='Velocity Profile')
plt.xlabel('Time (seconds)')
plt.ylabel('Velocity (m/s)')
plt.title('Vehicle Longitudinal Speed Control')
plt.legend()

# Plot the distance profile
plt.subplot(2, 1, 2)
plt.plot(time[:-1], distance_traveled, label='Distance Traveled', color='orange')
plt.xlabel('Time (seconds)')
plt.ylabel('Distance Traveled (meters)')
plt.legend()

plt.tight_layout()
plt.show()
````
